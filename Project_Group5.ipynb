#### 3. Machine Learning Algorithm

The dataset consists of **frequency** and **multiplicity**, where **multiplicity** is represented as categorical values in text format. To utilize this dataset effectively, we preprocessed **multiplicity** using a label encoder (include='object'). This step was necessary because most machine learning models cannot directly handle non-numerical, categorical data in its raw form.

To address the skewed distribution of the data, we applied the **StandardScaler** to transform the features into a standard normal distribution. This transformation ensures that the features have a mean of 0 and a standard deviation of 1, which helps improve the performance

We evaluated the performance of each model using **accuracy**, **F1 score**, **precision**, and **recall**, as these metrics offer different perspectives on model performance, making the evaluation more comprehensive:
- â **Accuracy**: Measures the overall correctness of predictions.
- **Precision**: The ratio of true positives to all predicted positives.
- â **Recall**: The proportion of true positives among actual positives.
- â **F1 Score**: The harmonic mean of precision and recall, balancing these two metrics.

To optimize the models, we employed **GridSearchCV**. These techniques helped identify the hyperparameters that yield the best performance. Furthermore, we utilized **cross-validation** to mitigate overfitting. Cross-validation splits the training set into multiple subsets for iterative training and validation, providing a robust estimate of model performance across different data partitions.

3-0. Pre-proccessing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
```

```python
X = Data.iloc[:, :-1]
y = Data.iloc[:, -1]

le = LabelEncoder()
for col in X.select_dtypes(include='object').columns:
    X[col] = le.fit_transform(X[col])
y = le.fit_transform(y)

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
results = {}
```

3-1. Random Forest

Random Forest generates multiple decision trees and combines their predictions to derive the final prediction:

1. Data Sampling: Bootstrap samples are created from the original dataset.
2. Decision Tree Creation: A decision tree is trained on each bootstrap sample. 
3. Prediction Aggregation: The predictions of individual trees are aggregated to produce the final output. (For classification problems, majority voting is used, while for regression problems, the average is taken.)

**Advantages**
- Random Forest generally exhibits flexible and stable performance, even for complex models.
- It was chosen because it is well-suited for this dataset, which contains non-standard data formats.

**Limitations**
- The training and prediction processes require high computational costs.
- There is a potential risk of overfitting for highly complex datasets.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Defining the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Defining Random Forest model
rf_model = RandomForestClassifier(random_state=42)
# cv - 10 loads cross-validation means we divide data into 10 parts to check the quality
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=10, n_jobs=-1, scoring='accuracy')
grid_search.fit(X_train, y_train)

print(f"Best Hyperparameters: {grid_search.best_params_}")

# Fitting the best Random Forest Model
best_rf_model = grid_search.best_estimator_
y_pred = best_rf_model.predict(X_test)

# Evaluating the performance
unique_classes = np.unique(np.concatenate((y_test, y_pred)))
target_names = le.inverse_transform(unique_classes)
print(classification_report(y_test, y_pred, target_names=target_names))

report_rf = classification_report(y_test, y_pred, output_dict=True)
results['Random Forest'] = {
    'Accuracy': report_rf['accuracy'],
    'Precision': report_rf['weighted avg']['precision'],
    'Recall': report_rf['weighted avg']['recall'],
    'F1 Score': report_rf['weighted avg']['f1-score']
}
```

3-2. SVM

Support Vector Machines (SVM) classify data by finding the optimal hyperplane that separates classes with the maximum margin. It focuses on maximizing the margin between the boundary and the closest support vectors (data points). 

**Advantages**
- performs well when the dataset has a large number of features since it focuses on finding the optimal decision boundary by maximizing the margin between classes
- the dataset includes mixed numerical and categorical features and therefore can handle non-linear decision boundaries effectively with kernels like RBF
- regularization parameter "C" elps control overfitting, especially in high-dimensional spaces  


**Limitations** 
- can be slow for large datasets
- sensitive to outliers and noise in the data
- requires features to be scaled since it relies on distance-based calculations

```python
from sklearn.svm import SVC
```

```python
# Defining the parameter grid for SVM
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 'auto', 0.001, 0.01, 0.1, 1],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
}

# Initializing SVM model
svm = SVC(random_state=42)

# GridSearchCV
grid_search = GridSearchCV(estimator=svm, param_grid=param_grid, cv=10, scoring='accuracy', verbose=1, n_jobs=-1)
grid_search.fit(X_train, y_train)

# Display the best parameters and corresponding score
print("Best parameters found:", grid_search.best_params_)
print("Best cross-validation accuracy:", grid_search.best_score_)

# Use the best estimator to make predictions
best_svm = grid_search.best_estimator_
y_pred = best_svm.predict(X_test)

# Evaluation
print("Classification Report:\n", classification_report(y_test, y_pred))

report_svm = classification_report(y_test, y_pred, output_dict=True)
results['SVM'] = {
    'Accuracy': report_svm['accuracy'],
    'Precision': report_svm['weighted avg']['precision'],
    'Recall': report_svm['weighted avg']['recall'],
    'F1 Score': report_svm['weighted avg']['f1-score']
}
```

3-3. LogisticRegression

1. Building a linear model, output - vector $z_j$, j correspond to a class
2. â Converting them into probabilities
3. â Predicted class is the class with the highest probability

**Advantages:**
- â Easy to implement.
- â Works well with small datasets.

**Limitations:**
- â Models can become imbalanced if one class dominates others.
- â Training separate models for each class can be computationally expensive.

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Define the parameter grid
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],  # Regularization types: l1(Lasso), l2 (Ridge), elasticnet(combined l1 and l2)
    'C': [0.01, 0.1, 1, 10, 100],  # Regularization strength (small - strict, big - few)
    'solver': ['lbfgs', 'saga', 'liblinear', 'newton-cg'],  # The way of solving
    'max_iter': [100, 200, 500]  # Iterations number for convergence
}

# Initializing
logreg = LogisticRegression()

# Set up GridSearchCV
grid_search = GridSearchCV(estimator=logreg, param_grid=param_grid, cv=10, scoring='accuracy', verbose=1, n_jobs=-1)
# cv - 10 loads cross-validation means we divide data into 10 parts to check the quality
# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Best parameters and corresponding score
print("Best parameters found:", grid_search.best_params_)
print("Best cross-validation accuracy:", grid_search.best_score_)

# Prediction
best_logreg = grid_search.best_estimator_
y_pred = best_logreg.predict(X_test)

# Evaluate performance
print("Classification Report:\n", classification_report(y_test, y_pred))
#Precision: correct predictions among all positive
#Recall: out of all actual positive examples, how many were found
#Accuracy: correct predictions
#F1-score: 2*precision*recall/(precision+recall) balance

report_logreg = classification_report(y_test, y_pred, output_dict=True)
results['Logistic Regression'] = {
    'Accuracy': report_logreg['accuracy'],
    'Precision': report_logreg['weighted avg']['precision'],
    'Recall': report_logreg['weighted avg']['recall'],
    'F1 Score': report_logreg['weighted avg']['f1-score']
}
```
